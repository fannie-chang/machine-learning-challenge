```python
import tensorflow as tf
import pandas as pd
import os
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("Resources/exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
target = df["koi_disposition"]
target_names = ["negative" , "positive"]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
data = df.drop("koi_disposition", axis =1)
feature_names = data.columns
data.head()
```

```python
# Split the data using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data


from sklearn.preprocessing import MinMaxScaler

X_scale = MinMaxScaler().fit(X_train)

X_train_scaled = X_scale.transform(X_train)
X_test_scaled = X_scale.transform(X_test)
```

# Train the Model



```python

from sklearn.neighbors import KNeighborsClassifier

train_scores = []
test_scores = []
for k in range (1,20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"K: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")


```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,20,2), train_scores, marker='o')
plt.plot(range(1,20,2), test_scores, marker='x')
plt.xlabel('k neighbors')
plt.ylabel('Testing accuracy Score')
plt.show()
```

```python
model_knn = KNeighborsClassifier(n_neighbors = 19)
model_knn.fit(X_train_scaled, y_train)

```

```python
print(f"Training Data Score: {knn.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {knn.score(X_test_scaled, y_test)}")
```

```python
predictions = knn.predict(X_test_scaled)
print(predictions)
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
from pprint import pprint
print('Parameters:\n')
pprint(model_knn.get_params())
```

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors': [ 7,9,11,15,17,19,21,23,25,27,29],
              'weights': ['uniform', 'distance'],
              'p': [1,2]}

grid = GridSearchCV(model_knn, param_grid, verbose=3 )
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
